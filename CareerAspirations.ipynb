```python
data["Your Current Country."].value_counts().plot(kind="bar")
plt.show()
```

```python
plt.hist(data["Your Gender"])
plt.show()
```

```python
Question1 = data['Which of the below factors influence the most about your career aspirations ?'].value_counts()
fig = px.pie(values=Question1.values, names=Question1.index, title='Factors influence the most about your career aspirations?')
fig.show()
```

```python
Question2 = data['Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.'].value_counts()
fig = px.pie(values=Question2.values,names=Question2.index, title='Would you definitely pursue a Higher Education outside of India by yourself ? ')
fig.show()
```

```python
Question3 = data['How likely is that you will work for one employer for 3 years or more ?'].value_counts()
fig = px.pie(values=Question3.values, names=Question3.index, title='How likely is that you will work for one employer for 3 years or more ?')
fig.show()
```

```python
Question4 = data['Would you work for a company whose mission is not clearly defined and publicly posted.'].value_counts()
fig = px.bar(Question4,title='Work for a company whose mission is not clearly defined and publicly posted?')
fig.show()
```

```python
Question5 = data['How likely would you work for a company whose mission is misaligned with their public actions or even their product ?'].value_counts()
fig = px.bar(Question5,title='Work for a company whose mission is misaligned with their public actions or even their product ?')
fig.show()
```

```python
Question6 = data['How likely would you work for a company whose mission is not bringing social impact ?'].value_counts()
fig = px.pie(values=Question6.values, names=Question6.index, title='Work for a company whose mission is not bringing social impact ? ')
fig.show()
```

```python
Question7 = data['What is the most preferred working environment for you.'].value_counts()
fig = px.pie(values=Question7.values, names=Question7.index, title='Most preferred working environment for you?')
fig.show()
```

```python
Question8 = data['Which of the below Employers would you work with.'].value_counts()
fig = px.pie(values=Question8.values, names=Question8.index, title=' Employers would you work with?')
fig.show()
```

```python
Question9 = data['Which type of learning environment that you are most likely to work in ?'].value_counts()
fig = px.pie(values=Question9.values, names=Question9.index, title='Learning environment that you are most likely to work in ?')
fig.show()
```

```python
Question11= data['What type of Manager would you work without looking into your watch ?'].value_counts()
fig = px.pie(values=Question11.values, names=Question11.index, title='Manager would you work with without looking into your watch ?')
fig.show()
```

```python
Question12= data['Which of the following setup you would like to work ?'].value_counts()
fig = px.bar(Question12, title='Setup you would like to work ?')
fig.show()
```
